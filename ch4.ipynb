```python
import matplotlib.pyplot as plt
import numpy as np
import pymc as pm
```

```python
# 尤度関数
def lh(p):
    return p ** 2 * (1-p) **3

# グラフの描画
p = np.arange(0.0,1.0,0.01)
plt.rcParams['figure.figsize'] = (6,4)
plt.plot(p,lh(p))
plt.xlabel('p')
plt.ylabel('likelihood')
# plt.title(f'尤度関数')
```
